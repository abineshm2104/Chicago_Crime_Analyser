```python
crime_set_df_missing = pd.DataFrame(crime_set_df.isnull().sum())
crime_set_df_missing = crime_set_df_missing.reset_index()
crime_set_df_missing.rename(columns = {'index': 'Features'}, inplace = True)
crime_set_df_missing.rename(columns = {0: 'Missing_Counts'}, inplace = True)
crime_set_missing_columns_lister = list(crime_set_df_missing['Features'])
print(crime_set_missing_columns_lister)
```

```python
for i in range(0, len(crime_set_missing_columns_lister), 1):
  percentage_missing = (crime_set_df_missing[crime_set_df_missing['Features']==crime_set_missing_columns_lister[i]]['Missing_Counts'][i] /crime_set_df.shape[0]) * 100
  if percentage_missing > 0.0:
    print(f'Missing Percentage of column {crime_set_missing_columns_lister[i]} is {percentage_missing} %')
```
